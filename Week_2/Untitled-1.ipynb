```python
import re, collections
with open('hw.txt', 'r') as file:
    array = []
    for line in file:
        array.append(str.lower(line.strip('\n')))
print(array)
```

```python
dict = {}
index = 0
for word in array:
  elements = re.split('[^a-z]', word)
  for element in elements:
    if element == '' or element in dict.values():
      continue
    else:
      dict[index] = element
      index +=1
      #print(word.split('[^a-z]', -1))
      #print(word)
print(dict)
```

```python
print(len(dict), len(array), dict)
```

```python
import numpy as np
```

```python
rows=[]
for element in array:
  #print(re.split('[^a-z]', element))
  row = []
  for key in dict:
    #print (dict[key], key, 'exists ', re.split('[^a-z]', element).count(dict[key]))
    row.append(re.split('[^a-z]', element).count(dict[key]))
  print(row)
  rows.append(row)
```

```python
import numpy as np
matrix = np.array(rows)
```

```python
matrix
```

```python
from scipy.spatial.distance import cosine
first_row = matrix[0]
counter = 0
answers = {}
for row in matrix:
    answers[counter] = cosine(first_row, row)
    counter +=1
print(answers)
```

```python
sorted_by_value = sorted(answers.items(), key=lambda kv: kv[1])
```

```python
sorted_by_value
```

```python
out_file = open('sub_1.txt', 'w')
ans_1 = sorted_by_value[1][0]
ans_2 = sorted_by_value[2][0]
result = str(ans_1)+' '+str(ans_2)
out_file.write(result)
out_file.close()
```
