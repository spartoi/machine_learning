```python
import re, collections
with open('hw.txt', 'r') as file:
    array = []
    for line in file:
        array.append(str.lower(line.strip('\n')))
print(array)
```

```python
dict = collections.defaultdict(lambda : -1) 
index = 0
for word in array:
  elements = re.split('[^a-z]', word)
  for element in elements:
    if element == '' or dict[element] > -1:
      continue
    else:
      dict[element] = index
      index +=1
      #print(word.split('[^a-z]', -1))
      #print(word)
print(dict)
```

```python
print(len(dict), len(array))
```

```python
import numpy as np
```

```python
rows=[]
for element in array:
  #print(re.split('[^a-z]', element))
  row = []
  for key in dict:
    print (dict[key], key)
    row.append(re.split('[^a-z]', element).count(key))
  rows.append(row)
```

```python
rows
```

```python
matrix = np.array(rows)
```

```python
matrix
```

```python
from scipy.spatial.distance import cosine
first_row = matrix[0]
counter = 0
answers = {}
for row in matrix:
    answers[counter] = cosine(first_row, row)
    counter +=1
print(answers)
```

```python
sorted_by_value = sorted(answers.items(), key=lambda kv: kv[1])
```
